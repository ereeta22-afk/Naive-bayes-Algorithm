```python
#-----Gaussian Naive -----------#
```

```python
Gaussian Naïve Bayes
Used for numerical data.
Calculates Mean and Variance.
Finds probability for each class.
Compares probabilities.
Chooses the highest probability.
Gives prediction.

Example: Height, Weight, Age
```

```python
from sklearn.naive_bayes import GaussianNB

X = data[["Height","Weight","Age","CGPA"]]
y = data["Gender"]

model = GaussianNB()
model.fit(X,y)

new_data = pd.DataFrame({
    "Height":[172],
    "Weight":[65],
    "Age":[21],
    "CGPA":[7.5]
})

result = model.predict(new_data)

print(result)
```

```python
#------------------multinomial naive-------------------------#
```

```python
Multinomial Naïve Bayes
Used for count/frequency data.
Counts feature occurrences.
Calculates probability for each class.
Compares probabilities.
Chooses the highest probability.
Gives prediction.

Example: Word counts in emails
```

```python
from sklearn.naive_bayes import MultinomialNB
data1=pd.read_csv("multinomial.csv")

X = data1[["Free_Count",
          "Win_Count",
          "Offer_Count",
          "Prize_Count",
          "Click_Count"]]

y = data1["Class"]
model = MultinomialNB()
model.fit(X, y)
new_data = pd.DataFrame({
    "Free_Count":[4],
    "Win_Count":[2],
    "Offer_Count":[1],
    "Prize_Count":[1],
    "Click_Count":[3]
})

result = model.predict(new_data)

print("Prediction:", result)
```

```python
#--------bernoulli  naive----------------------------------#
```

```python
Bernoulli Naïve Bayes - Very Simple Points
Uses 0 and 1 data.
Works with Yes/No values.
Learns from training data.
Uses Bayes Theorem.
Calculates probability.
Compares all classes.
Chooses the highest probability.
Gives the prediction.
```

```python
from sklearn.naive_bayes import BernoulliNB

data = pd.read_csv("bernoulli.csv")
X = data[["Fever", "Cough", "Headache"]]
y = data["Disease"]
model = BernoulliNB()
model.fit(X, y)

new_data = pd.DataFrame({
    "Fever": [1],
    "Cough": [1],
    "Headache": [0]
})
result = model.predict(new_data)

print("Prediction:", result)
```
